   The game is based on Harry Potter book series, in particular the 4th book of the series Harry Potter and the Goblet of Fire by J.K. Rowling. The purpose of the game is to find a way through the labyrinth to reach the Goblet and win the Triwizard Tournament. There is only one win function game_win at the end of the game, but wrong choices call for the game_over function which appears in the game quite often and results in failing to accomplish the task for the player. In order to win the game, the player has to make the right choices to reach the final stage of the game. The player is a fifth-year Slytherin student at Hogwarts School of Witchcraft and Wizardry and has been chosen to represent the school in the tournament.

   It has been assumed that the player has successfully accomplished the two previous tasks and for the last challenge, he or she needs to find their way through the labyrinth to find the Goblet of Fire and win the competition.

   The bug that has not been solved yet is creating within the code a counter that would take one attempt from the player's life-counter whenever they take a wrong path. When I add it to the code, the while loop stops working as I want, and I don't understand how to bring things back to normal. Instead, I end the game with the variable game_over that I declare every time the player takes a wrong path. Using a counter would save a lot of work by doing it automatically."
   
Rowling, J. K. (2002). Harry Potter #4: Harry Potter and the goblet of fire. Scholastic. 


```python
# function that runs when the player chooses to go back.
def go_back():
    print("\nYou decided not to participate in the Tournament.\n \nEveryone is laughing at you.\n \nBut at least you're safe, right?\n")
    print("\nThanks for playing The Triwizard Tournament game.\n")
    
# function that runs when the player made the wrong choice. It finishes the game.
def game_over():
    print("""\nMaybe next time you will not embarrass your school.\n \nShame on you.\n \nTry one more time if you're brave enough.\n""")

# function that runs when the player chooses all right choices and wins. It is called in the game only once.     
def game_win():
    print("\nUnfortunately, you couldn't resist.\n \nYou went straight to the Goblet and touched it.\n \nAs soon as your hand touched it, you were teleported to a scary dark cemetery.\n \nBut it seems like these creepy people in masks were waiting for someone else, a brown-haired boy with round glasses.\n \nWell, at least you won the competition!\n \nAvada Kedavra!\n")
    
# function that starts the game, prompts the player to enter their name and decide whether to enter the labyrinth or go back. 
# option go back ends the game.
def start_game():
    while True:
     print("""Welcome to the Triwizard Tournament!\n \nYou're a fifth-year Slytherin student at Hogwarts School of Witchcraft and Wizardry. \nYou have been chosen to represent your school in the Tournament.\n \nYou've successfully accomplished two previous tasks, and for the last challenge, you need to find your way through the labyrinth to find the Goblet of Fire and win the competition.\n \nBut first, introduce yourself:\n""")
     name = input("What is your name? ")
     print(f"\nHello, {name}!\n \nYou find yourself in front of the labyrinth.\n \nWhat would you like to do?")
     choice = input("\n\n1. Enter \n2. Go back\n")
     if "enter" in choice.lower() or "1" in choice.lower() or "one" in choice.lower():
         entrance()
         break
     elif "go" in choice.lower() or "back" in choice.lower() or "2" in choice.lower() or "two" in choice.lower():
         go_back()
         break 
     else:
         print("Invalid choice. Please try again.")  

 # function for the entrance (main) path. Both options don't result in game_over.
 # used for loop, lists           
def entrance():
    options = ["1. Go straight through the fog", "2. Turn right into the darkness"]
    for i in range(3):
        print("\nYou and your competitors entered the labyrinth.\n \nYou are surrounded by walls made of bushes, and now you need to decide which path to choose to proceed.\n")
        for option in options:
            print(option)
        choice = input()
        if "1" in choice or "straight" in choice.lower() or "fog" in choice.lower():
            fog()
            break
        elif "2" in choice or "turn" in choice.lower() or "right" in choice.lower() or "darkness" in choice.lower():
            darkness()
            break
        else:
            print("Invalid choice. Please try again.")
    
# function for the fog path    
# option explore ends the game
def fog():
    print("\nSeems like it's getting colder.\n \nYou hear creepy noises on the left.\n \nWhat would you do?")
    choice = input("1. Run!\n2.Carefully explore your surroundings. Maybe it's just a squirrel?")
    if "run" in choice.lower() or "1" in choice.lower() or "one" in choice.lower():
        bushes()
    elif "explore" in choice.lower() or "squirrel" in choice.lower() or "stay" in choice.lower() or "2" in choice.lower() or "two" in choice.lower():
        print("\nImmobulus! \nOh no!\n \nOne of the opponents just cast a spell on you.\n \nYou are unable to move until the rest of the tournament.\n \nHogwarts just lost the Tournament\n. \nBetter luck next time...\n")
        game_over()
    else:
        print("Invalid choice. Please try again.")    

 # function for the bushes path    
 # option run ends the game       
def bushes():
    print("\nSometimes it is better to be fast than curious.\n \nYou saved yourself. For now... \nBut something weird just touched your leg!\n \nQuickly, you need to...\n")
    choice = input("1. Faster legs are always better than any spells. Run, run, run... \n2. Incendio!")
    if "run" in choice.lower() or "legs" in choice.lower() or "1" in choice.lower() or "one" in choice.lower():
        print("\nSectumsempra!\n \nAnother player saw you running and thought you were a dangerous wild animal.\nYou died.\nBetter luck next time!\n")
        game_over()
    elif "Incendio" in choice.lower() or "2" in choice.lower() or "two" in choice.lower():
        fire()
    else:
        print("Invalid choice. Please try again.")    

# function for the fire path    
# option explore ends the game
# this finction is the last one in the game. 
# both options will be ending option
# but only option "hesitate" will lead to game_win
def fire():
    print("\nNasty bushes (which you just burnt!)\n \nSeems like someone is up to date with studying.\n \nThe Goblet of Fire just appeared in front of you!\n \nQuickly, go to it before your competitors will reach it first!\n")
    choice = input("1. Grab it! \n2. Hesitate. Something isn't right.")
    if "grab" in choice.lower() or "1" in choice.lower() or "one" in choice.lower():
        print("\nYou reached The Goblet.\n \nHowever, after touching it, you were teleported to San Francisco.\n \nYou are an MBAN student, studying python.\n \nDon't sleep in classes, magic isn't real.\n \nProfessor Chase is though.")
    elif "hesitate" in choice.lower() or "2" in choice.lower() or "two" in choice.lower():
        game_win()
    else:
        print("Invalid choice. Please try again.")
             
# funtion for the darkness path. 
# it is a dead end function. 
# Both options will result in game_over        
def darkness():
    print("""\nYou are in complete darkness, and it seems like your wand doesn't work properly.\n \nYou think you saw a light in front, but you are hesitant if you want to follow it or not.\n \nWhat would you do?""")
    choice = input("""1. Go towards the light. Maybe it's the way to the Goblet of Fire? \n2. Strange lights in the creepy labyrinth? No way you're going towards it. Turn left instead, maybe?""")
    if "light" in choice.lower() or "1" in choice.lower() or "one" in choice.lower() or "go" in choice.lower() or "goblet" in choice.lower():
        print("\nRing around the rosie.\n \nPocket full of posies.\n \nAshes, ashes.\n \nWe all fall down...\n \nA ghost of a dead contestant just appeared in front of you. \nYou got scared and sent a red emergency light to the sky.\n \nYou're out of the competition.\n \nDumbledore is ashamed.\n \nBetter luck next time!\n")
        game_over()
    elif "stay" in choice.lower() or "turn" in choice.lower() or "left" in choice.lower() or "2" in choice.lower() or "two" in choice.lower():
        print("""\nStaying in the darkness by yourself?\n \nNot a very smart choice...\n \nYou got a panic attack and withdrew yourself from the Tournament sending a red emergency light to the sky.\n \nHogwarts just lost the competition.\n \nSnape will be furious.\n""")
        game_over()
        
start_game()
```
